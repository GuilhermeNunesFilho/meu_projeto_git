```python
# Lista de compras

produto = []

while True:
  print("---Selecione o que deseja fazer!---")
  print("\nOpção 1: Adicionar Produto")
  print("Opção 2: Remover Produto")
  print("Opção 3: Listar Produtos")
  print("Opção 4: Sair")

  opcao = int(input("Digite a opção desejada: "))

  if opcao == 1:
    nome = input("Digite o nome do produto: ")
    produto.append(nome)

  elif opcao == 2:
    if len(produto) == 0:
      print("\nSua lista de Produtos está vazia!")
    else:
        try:
              escolha_usuario = input("Escolha o numero do item que deseja excluir: \n")
              posicao = int(escolha_usuario) - 1

                # Verifica se o número existe na lista
              if 0 <= posicao < len(produto):
                    remover = produto.pop(posicao)
                    print(f"O produto '{remover}' foi removido com sucesso!\n")
              else:
                    print("Esse número de item não existe na lista!\n")

        except ValueError:

              print("Erro: Por favor, digite apenas números inteiros válidos!\n")

  elif opcao == 3:
    if len(produto) == 0:
      print("\nSua lista de Produtos está vazia!")
    else:
      for indice, items in enumerate(produto):
        print(f"\n{indice +1} - {items}")

  elif opcao == 4:

    print("\nFinalizado, Até a próxima...")

    break


  else:
    print("\nOpção invalida!")
```
