```python
# Programa para atualizar um jogo
print("=========Placar da partida=========")
print("-"*35)
time1 = input("Quem vai joga?\n")
time2 = input("Contra quem?\n")
print("-"*35)
gol_time1 = 0
gol_time2 = 0

while True:
    try:
        print(f"PLACAR: {time1} = {gol_time1} VS {gol_time2} = {time2}")
        print("-"*35)
        print(f"1: Gol do time {time1}\n")
        print(f"2: Gol do time {time2}\n")
        print(f"3: Fim da Partida")
        print("-"*35)
        opcao = int(input("Escolha uma Opção?\n"))

        if opcao == 1:
          print(f"Goooooolll do {time1}")
          gol_time1 += 1
          print("-"*35)

        elif opcao == 2:
          print(f"Goooooolll do {time2}")
          gol_time2 += 1
          print("-"*35)

        elif opcao == 3:
          print(f"Final de partida entre {time1} vs {time2}")
          print("-"*35)
          print(f"------RESULTADO FINAL------\n{time1} = {gol_time1} / {gol_time2} = {time2}")

          break

        else:
          print("Opção invalida! Tente novamente usando uma dessas opções:\n1, 2 ou 3.")

    except ValueError:
      print("Error")

```
