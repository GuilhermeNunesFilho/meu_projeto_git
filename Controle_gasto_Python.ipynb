```python
valor_conta = []
nome_conta = []

while True:
    print("-" * 40)
    print("-------SEU CONTROLE DE GASTOS MENSAIS-----")
    print("-" * 40)
    print("1: Adicionar Conta")
    print("2: Mostrar Contas")
    print("3: Excluir Contas")
    print("4: Sair")
    print("-" * 40)

    try:
        opcao = int(input("Escolha uma opção:\n"))
    except ValueError:
        print("Por favor, digite apenas números para a opção!")
        continue

    print("-" * 40)

    if opcao == 1:
        conta = input("Nome da conta?\n")
        try:
            valor = float(input("Valor da Conta? (Use ponto para centavos):\n"))
            nome_conta.append(conta)
            valor_conta.append(valor)
            print("-" * 40)
            print(f"Conta '{conta}' no valor de R${valor:.2f} adicionada com sucesso!!")
        except ValueError:
            print("Erro: Digite o valor usando apenas números e ponto (Ex: 50.50).")

    elif opcao == 2:
        if not nome_conta:
            print("Nenhuma conta cadastrada ainda!")
        else:
            print("====== LISTA DE CONTAS ======")
            print("-" * 40)
            print("-------Conta---Valor-------")
            for indice, (list_conta, list_valor) in enumerate(zip(nome_conta, valor_conta), start=1):
                print(f"{indice}. {list_conta}: R$ {list_valor:.2f}")

    elif opcao == 3:
        if not nome_conta:
            print("Nenhuma conta cadastrada para excluir!")
        else:
            print("====== EXCLUIR CONTA ======")
            for indice, (list_conta, list_valor) in enumerate(zip(nome_conta, valor_conta), start=1):
                print(f"{indice}. {list_conta}: R$ {list_valor:.2f}")
            print("-" * 40)

            try:
                escolha = int(input("Digite o número da conta que deseja excluir:\n"))
                posicao = escolha - 1

                if 0 <= posicao < len(nome_conta):
                    conta_removida = nome_conta.pop(posicao)
                    valor_removido = valor_conta.pop(posicao)
                    print(f"Conta '{conta_removida}' excluída com sucesso!")
                else:
                    print("Número inválido! Essa conta não existe na lista.")
            except ValueError:
                print("Por favor, digite um número inteiro válido!")

    elif opcao == 4:
        print("Saindo do sistema... Até logo!")
        break
```
