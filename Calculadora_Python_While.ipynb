```python
# Calculadora

# Variáveis

sair = 1
opcao = 0

# Laço de repetição com menu e calculo do programa até que o usuário deseje sair

while sair == 1:

    num1 = int(input("Digite o primeiro número: "))
    num2 = int(input("Digite o segundo número: "))

    print("Escolha uma opção:")
    print("------------------")

    print("1 - Soma")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")

    opcao = int(input())
    print("------------------")

    if opcao == 1:
        print("Resultado: ",num1 + num2)
    elif opcao == 2:
        print("Resultado: ",num1 - num2)
    elif opcao == 3:
        print("Resultado: ",num1 * num2)
    elif opcao == 4:
        print("Resultado: ",num1 / num2)
    else:
        print("Opção Inválida")

# Perguntando se o usuário deseja sair

    sair = int(input("Deseja Continuar? \n1: Sim ou 2: Não?"))

# Tratando erro se caso for digitado algo diferente de 1 iy 2 em sair

    while sair != 1 and sair != 2:
        print("Opção Inválida")
        sair = int(input("Deseja Continuar? \n1: Sim ou 2: Não?"))

print("OK, Obrigado!!")
```
