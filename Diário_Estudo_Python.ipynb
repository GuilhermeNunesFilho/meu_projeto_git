```python
# Cofrinho digital

historico_nome = []
historico_valor = []

while True:
  print("=" * 35)
  print("--- 🐷 MEU COFRINHO DIGITAL ---")
  print("=" * 35)
  print("1:Depositar Dinheiro")
  print("2:Sacar Dinheiro")
  print("3:Ver Histórico")
  print("4:Sair")
  print("=" * 35)

  opcao = int(input("Escolha uma opção: "))
  print("=" * 35)
  if opcao == 1:
    motivo = input("Digite de onde veio este dinheiro:")
    try:
      print("=" * 35)
      print("Qual valor gostaria de Depositar?")
      print("=" * 35)
      valor = float(input("R$:"))

      if valor > 0:
        historico_nome.append(motivo)
        historico_valor.append(valor)
        print("=" * 35)
        print("Dinheiro Depositado com Sucesso!")
        print("=" * 35)
      else:
        print("Erro! O deposito precisa ser maior que '0'. ")
        print("=" * 35)
    except ValueError:
      print("Valor insuficiente! Tente novamente, por favor.")

  elif opcao == 2:
    try:
      print("=" * 35)
      print("Quantos gostaria de Sacar?")
      print("=" * 35)
      valor = float(input("R$:"))
      if valor > 0:
        historico_nome.append("Saque")
        historico_valor.append(-valor)
        print("=" * 35)
        print("Dinheiro Sacado com Sucesso!")
        print("=" * 35)
      else:
        print("Erro! O saque precisa ser maior que '0'. ")
        print("=" * 35)
    except ValueError:
      print("Valor insuficiente! Tente novamente, por favor.")

  elif opcao == 3:
    if len(historico_nome) == 0:
      print("Nem uma movimentação por enquanto!")
    else:
      print("=" * 35)
      print("\n--- EXTRATO DO COFRINHO ---")
      print("=" * 35)
      saldo_total = 0
      for indice, nome in enumerate(historico_nome):
        valor_movido = historico_valor[indice]

        if valor_movido > 0:
          print(f"{nome}: +R$ Entrada: {valor_movido:.2f}")
        else:
          print(f"{nome}: -R$ Saída: {-valor_movido:.2f}")

        saldo_total -= valor_movido
        print(f"Saldo Atual: R$ {-saldo_total:.2f}")
        print("=" * 35)

  elif opcao == 4:
    print("=" * 35)
    print("Até a próxima ;D")
    break

  else:
    print("=" * 35)
    print("Opção inválida. Tente novamente usando umas das opções validas!")
    print("=" * 35)

```
