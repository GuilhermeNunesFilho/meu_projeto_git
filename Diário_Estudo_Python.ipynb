```python
# Horas estudadas

materia = []
horas = []

# Laço de repetição do programa

while True:

  print("=" * 45)
  print("--------------Escolha uma Opção--------------")
  print("=" * 45)
  print("1: Adicionar")
  print("2: Mostrar")
  print("3: Sair")

  opcao = int(input("Digite a opção desejada: \n"))

  if opcao == 1:
    try:
      assunto = input("O que você quer estudar? \n")
      materia.append(assunto)
      tempo = float(input("Quanto tempo estuda no dia? \n"))
      horas.append(tempo)

      print("Conteúdo adicionado com sucesso!")

    except ValueError:
      print("Digite um valor válido!")

  elif opcao == 2:
    if len(materia) == 0:
      print("Lista Vazia!\n")

    else:
      print("=" * 45)
      print("---------Minhas diciplinas de Estudos--------")

      total_horas = 0.0
      for indice, item in enumerate(materia):

        tempo_item = horas[indice]

        print(f"{indice + 1} - {item}: {tempo_item} horas.\n")

        total_horas += tempo_item # soma as horas estudadas

      print(f"Total de tempo dedicado: {tempo} horas. Continue assim.\n")

  elif opcao == 3:
    print("Saindo!\n")
    break

  else:
    print("Opção invalida! Escolhaentre 1, 2 e 3.")
```
