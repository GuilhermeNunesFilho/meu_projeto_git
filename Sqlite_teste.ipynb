```python
# Usando o Python para interagir com o usuário

novo_nome = input('Digite um nome: ')
nova_idade = int(input('Digite uma idade: '))

# Inserindo dados na tabela,usuario, que foi fornecido pelo o usuário

cursor.execute("INSERT INTO usuario (nome, idade) VALUES (?, ?)", (novo_nome, nova_idade))

# Salvando as alterações no banco de dados

conexão.commit()

# Exibindo na tela a confirmação de inserção de dados

print(f"O usuário {novo_nome} foi adicionado com sucesso!")

# Mostrando a tabela

print("Tabela de Usuários:")

cursor.execute("SELECT * FROM usuario")

usuarios = cursor.fetchall()
for usuario in usuarios:
  print(f"ID: {usuario[0]}, Nome: {usuario[1]}, Idade: {usuario[2]}")
```
