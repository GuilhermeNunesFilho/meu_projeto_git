```python
#Importando o banco de dados Sqlite3
import sqlite3

conexao = sqlite3.connect('banco_de_dados_teste.db')
cursor = conexao.cursor()

#Criando uma tabela pessoas

cursor.execute("CREATE TABLE IF NOT EXISTS pessoas (id INTEGER, nome TEXT, idade INTEGER)")

#Adicionando dados na tabela pessoas

cursor.execute("INSERT INTO pessoas VALUES (1, 'João', 20)")
cursor.execute("INSERT INTO pessoas VALUES (2, 'Maria', 30)")
cursor.execute("INSERT INTO pessoas VALUES (3, 'Pedro', 40)")

#Salvando os dados no banco de dados

conexao.commit()

#Lendo as informações

cursor.execute("SELECT * FROM pessoas")

for linha in cursor.fetchall():
    print(linha)

#Fechando a conexão

conexao.close()
```
