```python
# Pergunte senha até acertar '1234'

tentativas = 0
while tentativas <= 3:
  senha = int(input('\nDigite a senha: '))
  tentativas += 1

  if senha == 1234:
    print("\nAcesso Liberado")
    break

  elif tentativas == 3:
    print("\nTentativas bloqueadas")
    break

  else:
    print("\nSenha Incorreta, tente novamente!")

print("\nFinalizado")
```
