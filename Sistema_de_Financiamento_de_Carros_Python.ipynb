```python
# Varredura de Clientes (Contador de Negocios)
print("====Cadastro====")
print("-" * 30)

# 1. Uma lista com vários scores de clientes diferentes
lista_scores = []
lista_nomes = []

for i in range(10):
    nome = input("Digite o nome do cliente:\n")
    score = int(input("Digite seu Score:"))
    lista_nomes.append(nome)
    lista_scores.append(score)

# 2. Crie aqui a sua variável para contar os negados (começa em 0)
contador_negados = 0
contador_juros_baixos = 0
# 3. O loop 'for' vai pegar um score por vez e colocar na variável 's'
for s in lista_scores:
    if s >= 700:
        contador_juros_baixos += 1
        print(f"Score {s}: Cliente com juros baixos!")
    elif s >= 500:
        print(f"Score {s}: Cliente com juros normais.")
    else:
        print(f"Score {s}: Financiamento Negado.")
        # SEU DESAFIO: O que você precisa colocar aqui para somar 1 no contador?
        contador_negados += 1

# 4. Exiba o valor final do contador na tela aqui embaixo (fora do loop)
print("-" * 30)
print(f"Total de clientes com juros reduzidos: {contador_juros_baixos}, parabéns")
print("-" * 30)
print(f"Total de clientes negados: {contador_negados}")
```
