```python
# Sistema de Financiamento de Carros

print("===Sistema de Financiamento de Carros===")
print("-"*40)
nome = input("Digite seu nome:\n")
idade = int(input("Digite sua idade:\n"))
score = int(input("Digite seu score:\n"))
print("-"*40)

if idade >= 18:
  if score >= 700:
    print("Financiamento Aprovado com juros baixo")
    print("-"*40)
    print(f"====|Cadastro Sr: {nome}|====")
    print("-"*40)
    print(f"Nome: {nome}\nIdade: {idade}\nScore: {score}")

  elif score >= 500:
    print("Financiamento Aprovado com juros normais")
    print("-"*40)
    print(f"====|Cadastro Sr: {nome}|====")
    print("-"*40)
    print(f"Nome: {nome}\nIdade: {idade}\nScore: {score}")

  elif score < 500:
    print("Financiamento não Aprovado, sentimos muito.")
    print("-"*40)
    print(f"====|Cadastro Sr: {nome}|====")
    print("-"*40)
    print(f"Nome: {nome}\nIdade: {idade}\nScore: {score} <<<< Score minimo de 500")

else:
  print(f"Obrigado por da preferencia ao nosso\nfinanciamento Sr@: {nome}, ficamos muito\nfelizes por sua escolha mas\ncom forme verificamos em seu cadastro ainda não\ntem os requisitos necessário para\ndar continuidade no financiamento.\nMuito Obrigado!!!")
  print("-"*40)
  print(f"====|Cadastro Sr: {nome}|====")
  print("-"*40)
  print(f"Nome: {nome}\nIdade: {idade} <<<< Idade não compativél\nScore: {score}")
```
