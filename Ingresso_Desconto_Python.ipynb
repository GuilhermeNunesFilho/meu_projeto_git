```python
ingresso = 40.0
estudante = True
apresentacao = input("Qual é o seu nome?\n")
idade = int(input("Quantos anos você tem:\n"))
estudante_ativo = input("Você é estudante?\n")
if estudante_ativo == "sim":
  estudante = True
else:
  estudante = False

if idade < 12:
  igresso_desc = round(ingresso * 0.5)
elif idade >= 60:
  igresso_desc = round(ingresso * 0.37 )
elif estudante == True:
  igresso_desc = round(ingresso * 0.8)
else:
  igresso_desc = ingresso

print(f"Que bom ter você aqui {apresentacao}\nHoje seu ingresso está apenas R${igresso_desc}")
```
