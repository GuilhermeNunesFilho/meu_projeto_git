```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
res = (nota1 + nota2 + nota3) / 3
if res >= 7:
  print(f"{res: .1f}")
  print("Aprovado")
elif res <= 6 and res >= 5:
  print(f"{res: .1f}")
  print("Recuperação")
else:
  print(f"{res: .1f}")
  print("Reprovado")
print("Fim.")
```
