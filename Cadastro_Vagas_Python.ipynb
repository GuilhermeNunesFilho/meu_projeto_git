```python
# Cadastro de Vagas
cadastrar_vagas = []
cadastrar_valou = []

while True:
    print("====Cadastrar Vagas====")
    print("-"*30)
    print("====Escolha uma opção====")
    print("-"*30)
    print("1: Cadastrar")
    print("2: Vagas")
    print("3: Sair")
    print("-"*30)

    try:
        opcao = int(input("O que deseja fazer?\n"))
    except ValueError:
        print("Digite uma opção válida!")
        print("-"*30)
        continue

    # --- OPÇÃO 1: CADASTRAR ---
    if opcao == 1:
        vaga_nome = input("Digite o nome da vaga: ")
        try:
            # Usamos float() para salvar como número e permitir contas matemáticas
            vaga_valor = float(input("Digite o valor da vaga: R$ "))

            cadastrar_vagas.append(vaga_nome)
            cadastrar_valou.append(vaga_valor)
            print("-"*30)
            print("Vaga cadastrada com sucesso!")
            print("-"*30)
        except ValueError:
            print("Erro: Digite apenas números válidos para o valor da vaga!")
            print("-"*30)

    # --- OPÇÃO 2: LISTAR E CALCULAR MÉDIA ---
    elif opcao == 2:
        print("-"*30)
        print("Vagas cadastradas:")
        print("-"*30)

        if len(cadastrar_vagas) == 0:
            print("Nenhuma vaga cadastrada!")
            print("-"*30)
        else:
            # Mostra todas as vagas cadastradas
            for i in range(len(cadastrar_vagas)):
                print(f"{i+1}: {cadastrar_vagas[i]} - R$ {cadastrar_valou[i]:.2f}")
            print("-"*30)

            # O cálculo da média fica protegido aqui dentro, após listar tudo
            media = sum(cadastrar_valou) / len(cadastrar_valou)
            print(f"Média dos valores das vagas: R$ {media:.2f}")
            print("-" * 30)

    # --- OPÇÃO 3: SAIR ---
    elif opcao == "3" or opcao == 3:
        print("-"*30)
        print("Saindo do programa...")
        print("-"*30)
        break

    # --- OPÇÕES INVÁLIDAS ---
    else:
        print("Opção inválida!")
        print("-"*30)
```
